```python
import Levenshtein
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'col1':['hi budy','hello','code','programming'],
                  'col2':['hey budy','hello','progra','coding']})

def text_similarity(str1,str2):
    distance = Levenshtein.distance(str1,str2)
    similarity = 100 - ((distance / max(len(str1),len(str2)))) * 100
    return similarity

# str1 = "hello world"
# str2 = "hellow worlds"


df['similarity'] = df.apply(lambda x:text_similarity(x['col1'],x['col2']),axis=1)
print(df)
```
